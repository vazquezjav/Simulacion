## Dise√±e y desarrolle un modelo y/o script que permita simular el siguiente caso real:
 En base a los datos del siguiente link https://educacion.gob.ec/wp-content/uploads/downloads/2012/08/AZUAY11.pdf, genere una simulaci√≥n del ingreso de los estudiantes, para ello debemos escoger un establecimiento y en base a los docentes y estudiantes modelar el reingreso a las aulas.
- Solo se va a tener en cuenta uno de los planteles educativos (escuela, colegio, universidad dentro del Azuay).
- Se tiene un promedio que el 90% de los docentes han sido vacunados y pueden realizar el proceso de ingreso en cada uno de los cursos.
- Dentro del proceso se tiene que alrededor del 5% - 10% de los estudiantes no podr√°n asistir debido a no presentar la vacuna/enfermedades.
- Los estudiantes solo pertenecen a una sola entidad educativa al igual que los docentes.
- Se va  a tener un periodo de prueba de un mes, posterior a ello se realiza al azar al 10% de estudiantes una prueba PCR para validar que no est√©n contagiados.
- De la ultima el 2% de los estudiantes dan positivo por lo que se cierra el curso completo.
- Los estudiantes asisten cada semana y estos est√°n en un horario de 6 horas ya sea diurno o nocturno.
- Tienen un receso 30 minutos dentro del establecimiento en donde se concentran todos los estudiantes y es un foco de contagio del 2%.

```python
import simpy
import random 
import math 
import datetime as dt
from datetime import datetime
import matplotlib.pyplot as plt
```

### Datos -  Institucion Padre Carlos de Bellavista


* Docentes = 37
* Estudiantes = 924

### Variables de Interes

```python
DOCENTES = 37  # Cantidad de docentes de la institucion analizada
ESTUDIANTES = 924  # Estudiantes matriculados en el ultimo a√±o lectivo de la institucion 
SEMANAS = 6       
HORAS_CLASE = 6 # horas diarias de clase 
TIEMPO_SIMULACION = SEMANAS * 5 * HORAS_CLASE * 60
FECHA_ACTUAL = datetime.now().strftime('%d/%m/%Y')
```

### Variables auxiliares 


```python
cont = {0:1}
estado_estudiantes={}
inicia_dia=[]
termina_dia=[]
docentes_estado=[]
```

```python
class Institucion(object):
    def __init__(self, env, docentes_habilitados):
        self.env = env
        self.aulas = simpy.Resource(env, capacity= docentes_habilitados)
        self.institucion = simpy.Container(env,init=0, capacity=10000)
        self.clase = False
        self.receso = False
        #self.estudiantes_clase = 0
        self.cont = 0
        self.monitoreo_clase = env.process(self.clases())
        
        #self.rmonitoreo_receso = env.process(self.recesoEstudiante())
    def logs_procesos(self,icon,nombre, accion, hora):
        print(" %s nombre: [%s] accion:  [%s] hora:  [%s] "%(icon,nombre, accion, hora))
        
    def clases(self):
        while True:
            if not self.receso:
                
                if self.cont == 0:
                    
                    self.logs_procesos("üë®‚Äçüè´", '',"empieza las clases", '7:15 am')
                    self.clase = True
                    yield self.env.timeout(165)
                    self.receso = True
                    self.logs_procesos("<--üë®‚Äçüè´", '',"Termina las clases a las.", '10:00')
                    self.clase = False
                    yield env.process(self.recesoEstudiante())
                    self.receso = False
                else:
                    self.logs_procesos("üë®‚Äçüè´", '',"empieza las clases despues del receso", '10:30')
                    yield self.env.timeout(165)
                    for key in estado_estudiantes:
                        if estado_estudiantes[key]=='clase':
                            estado_estudiantes[key]='termina jornada'
                    self.cont = 0
                
                
    def recesoEstudiante(self):
        self.logs_procesos("\o/", '',"Empieza receso de 30 minutos ", '10:30')
        yield self.env.timeout(30)
        self.receso = False
        self.clase = True
        self.cont = 1
        for i in estado_estudiantes:
            if estado_estudiantes[i]=='clase':
                if random.randint(1,1000) >5:
                    estado_estudiantes[i]='clase'
                else:
                    self.logs_procesos("‚ö†Ô∏è", i,"contagiado en el receso", '')
                    estado_estudiantes[i]='contagiado receso'
                    self.institucion.get(1)
        if random.randint(0,100)<=3:
            docentes_estado.append(random.randint(0,DOCENTES))  
```

```python
class Estudiante():
    def __init__(self, env, nombre, institucion):
        self.env = env
        self.nombre = nombre
        self.institucion = institucion
        #self.llegada = env.process(self.procesos_estudiantes())
        self.numero_estudiante = 0 
        
    def logs_procesos(self,icon,nombre, accion, hora):
        print(" %s nombre: [%s] accion:  [%s] hora:  [%d] "%(icon,nombre, accion, hora))
        

    def proceso_llegar(self,nombre):
        #self.logs_procesos("--->",nombre," llega a la institucion",self.env.now)
        estado_estudiantes[nombre]='llega'
        self.numero_estudiante +=1
        
    def comprobar_enfermedades(self):
        for i in estado_estudiantes:
            if estado_estudiantes[i]=='llega':
                if random.randint(1,1000) > 5:
                    #self.logs_procesos("üõÇ", i, "pasa al aula de clases", self.env.now)
                    estado_estudiantes[i]='clase'
                    institucion.institucion.put(1)
                else:
                    self.logs_procesos("‚ùå", i,"presenta problemas de salud no acude a clases", self.env.now)
                    estado_estudiantes[i]='enfermedad'
                    
                    
class Aula():
    def __init__(self, env,nombre, estado_estudiantes):
        self.env = env
        self.nombre = nombre
        self.aulas = simpy.Resource(env, capacity=DOCENTES)
        self.numero = 0
        self.proce= env.process(self.entrarAula(estado_estudiantes))
        
    def entrarAula(self,estado_estudiantes):
        while True:
            for i in estado_estudiantes:
                if estado_estudiantes[i]=='clase':
                    estado_estudiantes[i]='aula'
                    self.numero +=1
                    yield self.env.timeout(0)
```

```python
def clase(env,institucion, tiempo_asistencia,estudiante, dia):
    if dia ==1:
        for i in range(ESTUDIANTES):
            nombre = 'estudiante '+str(i+1)
            estudiante.proceso_llegar(nombre)
            #institucion.institucion.put(1)
        estudiante.comprobar_enfermedades()
       # aula = [Aula(env,'Aula %d'%i, estado_estudiantes) for i in range (DOCENTES)]
       # for es in aula:
       #     print("[%s] ha vacunado a [%d] personas"%(es.nombre,es.numero))
    else:
        for i in estado_estudiantes:
            if estado_estudiantes[i] =='termina jornada':
                estudiante.proceso_llegar(i)
        estudiante.comprobar_enfermedades()
    print('Inicia el dia %d con %s estudiantes hora %d'%(dia,institucion.institucion.level, env.now))
    inicia_dia.append(institucion.institucion.level)
    yield env.timeout(0)


def generarLlegada(env, institucion, docentes):
    print("Docentes habilitados para dar clase: ", docentes)
    estudiante = Estudiante(env,'', institucion)
    dia =1
    tiempo_asistencia = HORAS_CLASE *60
    while True:

        fecha_dia = datetime.strftime(datetime.strptime(FECHA_ACTUAL, '%d/%m/%Y')+dt.timedelta(days=dia-1), '%d/%m/%Y')
        print('-'*30,"Dia %d"%dia,'-'*15, fecha_dia,'\n')
        
        env.process(clase(env, institucion,tiempo_asistencia, estudiante, dia))
        #env.process(clase2(env, institucion,tiempo_asistencia, dia, docentes))
        yield env.timeout(tiempo_asistencia+1)
        print('Termina el dia %d con %s estudiantes '%(dia,institucion.institucion.level))
        termina_dia.append(institucion.institucion.level)
        institucion.institucion.get(institucion.institucion.level)
        dia +=1
        
    
```

```python
docentes_habilitados = math.ceil(DOCENTES*0.9)

env = simpy.Environment()
institucion = Institucion(env, docentes_habilitados)
env.process(generarLlegada(env,institucion, docentes_habilitados))
env.run(until=TIEMPO_SIMULACION)
```

### Termina mes de prueba

Termina el mes de prueba y se procede a escojer al azar el 10% del total de los estudiantes que se encuentran en clases, de este porcentaje el 2% dan positivo a la prueba PCR 

```python
def pruebaPCR(estudiantes):
    asistentes=[key for key in list(estudiantes.keys()) if estudiantes[key]=='clase']
    val=round(len(asistentes)*0.1)
    nombres=[]
    lista_estudiantes=[]
    for i in range(val):
        #est = random.randint(0,len(asistentes))
        est=random.choice(asistentes)
        
        if est  in lista_estudiantes:
            est=random.choice(asistentes)
        lista_estudiantes.append(est)
        
        if random.randint(0,50) <=2:
            #print("Positivo a la prueba PCR ", asistentes[est])
            nombres.append(est)
            print('‚ö†Ô∏è'*2,'%s elegido  para realizarse la prueba'%est,' Resultado: Positivo','‚ö†Ô∏è'*2)
        else:
            print('%s elegido  para realizarse la prueba'%est,' Resultado: Negativo')
        
    return nombres
pcr_nombre=pruebaPCR(estado_estudiantes)
pcr_nombre
```
