# Modelo 2D

```python
import pygame
import random 
import math 
import time 
import matplotlib.pyplot as plt


RADIUS = 5
(WIDTH, HEIGHT) = (900,700)
TEXT_BOX_HEIGHT = 100 
SPEED = 2
DIA=5 # SEGUNDOS 

WINDOWS_COLOR = (255,255,255)
WINDOWS_TITLE = 'SIMULACION ESCUELA'

NUM_STUDENTS = 924
DOCENTS_=37

pygame.init() # activacion libreria pygame

#creacion de la ventana 
screen = pygame.display.set_mode((WIDTH,HEIGHT + TEXT_BOX_HEIGHT))
pygame.display.set_caption(WINDOWS_TITLE)
screen.fill(WINDOWS_COLOR)


class Student():
    def __init__(self, x, y, angle, color):
        self.angle = angle
        self.x = x
        self.y = y
        self.speed = random.random() # caracteristica del estudiante 
        self.current_speed = self.speed
        self.collision_time = 0
        self.collided = False
        
        self.radius = RADIUS
        self.color = color
        self.status = 'sano'
        self.onset = 0
        
    def display(self,):
        pygame.draw.circle(screen, self.color,(int(self.x), int(self.y)), self.radius)
        
    def move(self):
        self.x += math.cos(self.angle)*self.current_speed*SPEED
        self.y += -math.sin(self.angle)*self.current_speed*SPEED
        
    # comprobar si necesita rebotar en la pared 
    def bounce(self):
        if self.x > WIDTH-self.radius:
            self.x = 2 *(WIDTH-self.radius) - self.x
            self.angle = math.pi-self.angle
            
        elif self.x < self.radius:
            self.x = 2*self.radius -self.x
            self.angle = math.pi -self.angle
            
        if self.y > HEIGHT-self.radius:
            self.y = 2 * (HEIGHT-self.radius) -self.y
            self.angle =  -self.angle
            
        elif self.y < self.radius:
            self.y = 2 *self.radius-self.y
            self.angle = - self.angle
    
    # infectar particula 
    def infect(self):
        if self.status !='infectado':
            num = random.randint(1,100)
            if num <= 10:
                self.color = (255,0,0)
                self.status = "infectado"
                self.onset = time.time()
                
    def infect_init(self,particles):
        for i in particles:
            i.color=(255,0,0)
            i.status='infectado'
    
            
# verificar si dos particulas estan cercas y ajustar centros 

def collide(particles):
    for i in range(0,len(particles) - 1):
        target = particles[i]
        
        for j in range(i+1, len(particles)):
            other = particles[j]
            diff_x = target.x - other.x
            diff_y = target.y - other.y
            distance = math.sqrt(diff_x**2 + diff_y**2)
            
            if distance < RADIUS * 2:
                
                # ajustar direcciones, cambiar velocidades 
                temp_angle = target.angle
                tem_speed = target.current_speed
                target.angle = other.angle
                target.current_speed = other.current_speed
                other.angle = temp_angle
                other.current_speed = tem_speed
                target.collision_time = time.time()
                other.collision_time = time.time()
                target.collided = True
                other.collision_time = True
                
                if other.status =='infectado' or target.status=='infectado':
                    other.infect()
                    target.infect()
                    
                if distance != 0:
                    target.x = other.x+diff_x/distance*2*RADIUS
                    target.y = other.y+diff_y/distance*2*RADIUS
                    
def main():
    students = []
    for i in range(NUM_STUDENTS):
        x = random.randint(RADIUS,WIDTH-RADIUS)
        y = random.randint(RADIUS, HEIGHT - RADIUS)
        angle = random.uniform(0,2*math.pi)
        color=(0,0,255)
        students.append(Student(x,y,angle, color))
     
    students[0].infect()
    students[0].infect_init(students[0:4])
    # inicializar estados contadores 
    healthy = 0
    infected = 1
    
    healthy_counts = []
    infected_counts = []
    
    running = True
    
    
    #docentes 
    docents = []
    for i in range(DOCENTS_):
        x = random.randint(RADIUS,WIDTH-RADIUS)
        y = random.randint(RADIUS, HEIGHT - RADIUS)
        angle = random.uniform(0,2*math.pi)
        color=(255,205,150)
        docents.append(Student(x,y,angle, color))
    cont_dias = 0
    cont = 0 
    while running:
        delay = pygame.time.delay(1)
        
        cont +=1
        if cont%15==0:
            print('dia %d'%cont_dias)
            cont_dias +=1
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                running = False
        collide(students)
        collide(docents)
        healthy = 0
        infected = 0
        docent_health = 0
        for p in students:
            p.move()
            p.bounce()
            p.display()
            
            if p.status =='sano':
                healthy +=1
            elif p.status =='infectado':
                infected +=1
        
        for p in docents:
            p.move()
            p.bounce()
            p.display()
            
            if p.status =='sano':
                docent_health +=1
            elif p.status =='infectado':
                infected +=1
        healthy_counts.append(healthy)
        infected_counts.append(infected)
        
        #creacmos fuenta para el texto 
        font = pygame.font.Font(pygame.font.get_default_font(),15)
        
        text = font.render('------ DIA ------ '+str(cont_dias)+'                            ESTUDIANTES MATRICULADOS:     '+str(NUM_STUDENTS), True, (255,0,255))
        screen.blit(text,(10,HEIGHT))
        text = font.render('Estudiantes Sanos: '+str(healthy), True, (0,0,255))
        screen.blit(text,(10,HEIGHT+20))
        text = font.render('Infectados: '+str(infected), True,(255,0,0))
        screen.blit(text, (10,HEIGHT+40))
        text = font.render('Docentes Sanos: '+str(docent_health), True,(255,205,150))
        screen.blit(text, (10,HEIGHT+60))
        text = font.render('ESCUELA PADRE CARLOS BELLAVISTA ', True,(255,0,0))
        screen.blit(text, (250,HEIGHT+80))
        
        # actualizar ventana
        pygame.display.update()
        screen.fill(WINDOWS_COLOR)

# ejecutar modelo 3d

main()
```
