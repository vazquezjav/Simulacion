```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
vacunas = pd.read_csv('vacunas.csv')
plan = pd.read_csv('plan_vacunarse.csv')
total=pd.read_csv('total_vacunas.csv')
total.head()
```

# Graficas y reportes del total de personas vacunadas


```python
plt.figure(figsize=(15,8))
plt.plot( total['fecha'],total['dosis_total'],'.-',total['fecha'],total['primera_dosis'],'o-',total['fecha'],total['segunda_dosis'],'*-')
plt.legend(['Dosis Total','Primera','Segunda'])
plt.xticks(total.fecha[::5].tolist())
plt.title('Distribucion Dosis')
```

```python
plt.figure(figsize=(15,8))
plt.plot(plan['fecha'], plan['primera_dosis'],'-*r')
plt.xticks(plan.fecha[::5].tolist(),rotation=50)
plt.ylabel('Dosis')
plt.xlabel('Dias')
plt.title('Distribucion aplicacion primera dosis')
```

en el grafico podemos observar cual a ido siendo las dosis de vacunas aplicadas a las personas en un solo dia, se aprecia claramente como a ido variando este valor a lo largo de los dias

## Generar un grafico por fabricante de vacuna 

```python
fizer = vacunas.loc[vacunas['vaccine'] == 'Pfizer/BioNTech']['total'].sum()
sinovac = vacunas.loc[vacunas['vaccine'] == 'Sinovac']['total'].sum()
oxford = vacunas.loc[vacunas['vaccine'] == 'Oxford/AstraZeneca']['total'].sum()

plt.figure(figsize=(8,5))
etiquetas = ['Pfizer/BioNTech', 'Sinovac','Oxford/AstraZeneca']
colores = ['#EA2FC8', '#B0CE31','#a94d39']
plt.pie([fizer, sinovac,oxford], labels=etiquetas, colors=colores, autopct='%.2f %%')
plt.title('Vacunas Fabricante')
```

## #GRAFICOS POR MES LLEGADO VACUNAS

```python

fizer2 = vacunas.loc[vacunas['vaccine'] == 'Pfizer/BioNTech']
fizer2['arrived_at'] = pd.to_datetime(fizer2['arrived_at'])
meses=[]
fizer2['arrived_at']=pd.to_datetime(fizer2['arrived_at'].astype(str), format='%Y-%m-%d')
enero_fizer=fizer2[(fizer2['arrived_at'] >= '2021-01-01') & (fizer2['arrived_at'] <= '2021-01-30')]
meses.append(enero_fizer['total'].sum())

meses.append(fizer2[(fizer2['arrived_at'] >= '2021-02-01') & (fizer2['arrived_at'] <= '2021-02-28')]['total'].sum())
meses.append(fizer2[(fizer2['arrived_at'] >= '2021-03-01') & (fizer2['arrived_at'] <= '2021-03-31')]['total'].sum())
meses.append(fizer2[(fizer2['arrived_at'] >= '2021-04-01') & (fizer2['arrived_at'] <= '2021-04-30')]['total'].sum())
meses.append(fizer2[(fizer2['arrived_at'] >= '2021-05-01') & (fizer2['arrived_at'] <= '2021-05-31')]['total'].sum())
etiquetas=['Enero','febrero','marzo','abril', 'mayo']
plt.bar(etiquetas, meses, align='center', alpha=0.5)
plt.xlabel('meses')
plt.title('Pfizer/BioNTech')
plt.ylabel('Dosis')
```

```python
#SINOVAC
fizer2 = vacunas.loc[vacunas['vaccine'] == 'Sinovac']
#fizer2['arrived_at'] = pd.to_datetime(fizer2['arrived_at'])
fizer2['arrived_at']=pd.to_datetime(fizer2['arrived_at'],format='%d/%m/%Y')
meses_sino=[]
meses_sino.append(fizer2[(fizer2['arrived_at'] >= '2021-01-01') & (fizer2['arrived_at'] <= '2021-01-30')]['total'].sum())
meses_sino.append(fizer2[(fizer2['arrived_at'] >= '2021-02-01') & (fizer2['arrived_at'] <= '2021-02-28')]['total'].sum())
meses_sino.append(fizer2[(fizer2['arrived_at'] >= '2021-03-01') & (fizer2['arrived_at'] <= '2021-03-30')]['total'].sum())
meses_sino.append(fizer2[(fizer2['arrived_at'] >= '2021-04-01') & (fizer2['arrived_at'] <= '2021-04-30')]['total'].sum())
meses_sino.append(fizer2[(fizer2['arrived_at'] >= '2021-05-01') & (fizer2['arrived_at'] <= '2021-05-31')]['total'].sum())

plt.bar(etiquetas, meses_sino, align='center', alpha=0.5, color='g')
plt.xlabel('meses')
plt.title('Sinovac')
plt.ylabel('Dosis')
```

```python
#Oxford/AstraZeneca
oxford = vacunas.loc[vacunas['vaccine'] == 'Oxford/AstraZeneca']
oxford['arrived_at'] = pd.to_datetime(oxford['arrived_at'])
oxford['arrived_at']=pd.to_datetime(oxford['arrived_at'].dt.strftime("%Y/%m/%d"))
meses_oxford=[]
meses_oxford.append(oxford[(oxford['arrived_at'] >= '2021-01-01') & (oxford['arrived_at'] <= '2021-01-30')]['total'].sum())
meses_oxford.append(oxford[(oxford['arrived_at'] >= '2021-02-01') & (oxford['arrived_at'] <= '2021-02-28')]['total'].sum())
meses_oxford.append(oxford[(oxford['arrived_at'] >= '2021-03-01') & (oxford['arrived_at'] <= '2021-03-30')]['total'].sum())
meses_oxford.append(oxford[(oxford['arrived_at'] >= '2021-04-01') & (oxford['arrived_at'] <= '2021-04-30')]['total'].sum())
meses_oxford.append(oxford[(oxford['arrived_at'] >= '2021-05-01') & (oxford['arrived_at'] <= '2021-05-31')]['total'].sum())

plt.bar(etiquetas, meses_oxford, align='center', alpha=0.5,color='r')
plt.xlabel('meses')
plt.title('Oxford/AstraZeneca')
plt.ylabel('Dosis')
```

Analizando las diferentes graficas, claramente la vacuna de Pfizer es la que continuamente a estado recibiendo ecuador en los ultimos 5 meses, aunque no es con la que mas dosis cuenta, claramente la vacuna Sinovac es la que mas dosis a llegado a Ecuador seguida de la Oxford que llegaron unicamente en los meses de marzo y abril

```python
plt.figure(figsize = (15,5))
plt.subplot(1, 3, 1)
plt.bar(etiquetas, meses, align='center', alpha=0.5)
plt.xlabel('meses')
plt.title('Pfizer/BioNTech')
plt.ylabel('Dosis')

plt.subplot(1, 3, 2)
plt.bar(etiquetas, meses_sino, align='center', alpha=0.5, color='g')
plt.xlabel('meses')
plt.title('Sinovac')
plt.ylabel('Dosis')

plt.subplot(1, 3, 3)
plt.bar(etiquetas, meses_oxford, align='center', alpha=0.5,color='r')
plt.xlabel('meses')
plt.title('Oxford/AstraZeneca')
plt.ylabel('Dosis')
```
