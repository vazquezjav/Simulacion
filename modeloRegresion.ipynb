```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
covid_df = pd.read_csv('../Recursos-Tareas/CovidEcuador.csv')
covid_df = covid_df.fillna(0)
covid_df['date']=pd.to_datetime(covid_df["date"]).dt.strftime("%Y%m%d")
covid_df
```

```python

x = covid_df[['date', 'new_cases', 'total_deaths', 'new_deaths', 'new_tests','total_tests','total_vaccinations']]

#para separar nuestra variable dependiente de la independiente, haremos lo siguiente:

y = covid_df['total_cases']

x.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=.2, random_state=42)
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression 
modelo = LinearRegression()
modelo.fit(X_train, y_train)
```

```python
print('w = ' + str(modelo.coef_) + ', b = ' + str(modelo.intercept_))
```
